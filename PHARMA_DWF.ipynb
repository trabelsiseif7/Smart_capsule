```python
Dim_Mutuelle= Dimension( name='Dim_Mutuelle',
                         key='Mutual_PK',               
                         attributes=['CodeMutuelle','NomMutuelle','PriseEnCharge'], 
                         lookupatts=(['CodeMutuelle']),
                         idfinder=None,
                         defaultidvalue=-1
                        )
```

```python
dic={}
for j in Mutuelle_SA.index: 
    for i in Mutuelle_SA:
        dic[i]=Mutuelle_SA[i][j]
    
    
      
    key = Dim_Mutuelle.lookup(dic)
   
    if key==-1:
        Dim_Mutuelle.insert(dic)
    else:
        
        Dim_Mutuelle.ensure(dic)
    
conn4.commit()
```
